```python
import os
import sys
import psycopg2

file = open('join_date.txt','r')
join_date = file.read().strip()
file.close()

try:
    connection = psycopg2.connect(
        database = sys.argv[1],
        user = os.environ.get('PGUSER'),
        password = os.environ.get('PGPASSWORD'),
        host = os.environ.get('PGHOST'),
        port = os.environ.get('PGPORT')
    )
    cursor = connection.cursor()
    
    query = """SELECT e.name, e.joining_date FROM employees e WHERE e.joining_date < %s"""
    
    cursor.execute(query, (join_date,))

    results = cursor.fetchall()
    for result in results:
        print(f"{result[0]}, {result[1]}")

    cursor.close()

except Exception as e:
    print(f"Error: {e}")

finally:
    if connection:
        connection.close()
```

```python
import os
import sys
import psycopg2

with open('course_name.txt', 'r') as file:
    course_name = file.read().strip()

try:
    connection = psycopg2.connect(
        database=sys.argv[1],
        user=os.environ.get('PGUSER'),
        password=os.environ.get('PGPASSWORD'),
        host=os.environ.get('PGHOST'),
        port=os.environ.get('PGPORT')
    )
    cursor = connection.cursor()
    
    query = """
    SELECT s.name 
    FROM students s 
    JOIN enrollments e ON e.student_id = s.student_id 
    JOIN courses c ON c.course_id = e.course_id 
    WHERE c.name = %s
    """
    cursor.execute(query, (course_name,))
    
    results = cursor.fetchall()
    for result in results:
        print(result[0]) 

    cursor.close()

except Exception as e:
    print(f"Error: {e}")

finally:
    if connection:
        connection.close()
```
